```python
import matplotlib.pyplot as plt;
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#create a line plot
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("x vs f(x^2)")
```

```python
#creatig a customized line plot
plt.plot(x,y,color='blue',linestyle='-')
```

```python
plt.plot(x,y,color='red',linestyle='--')
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=8)
```

```python
#multiple plots

y1=x

plt.Figure(figsize=(9,5))

plt.subplot(1,2,1)#(row,coloumn,which col)
plt.plot(x,y,color='green')
plt.title("plot1")

plt.subplot(1,2,2)#(row,coloumn,which col)
plt.plot(x,y1,color='red')
plt.title("plot2")

```

```python
#Bar plot
category=['A','B','C','D','E']
val=[1,2,3,4,5]

#create a bar plot
plt.bar(category,val,color='green')
plt.title("bar graph 1")
```

```python
#histogram
#->used to represent the distribution of a dataset
#->they divide tha data bins and count the no of data points in each bin
d=[1,2,3,4,54,5,21,1,2,4,9,5,5]
plt.hist(d,bins=5,color='orange',edgecolor='black')
```

```python
#scatterplot
plt.scatter(x,y,color='green',marker='v')
```

```python
#pie charts
labeld=['A','B','C','D','E']
sizes=[30,40,10,2,8]
colors=['green','blue','orange','yellow','red']
explode=(0.2,0,0,0,0)

plt.pie(sizes,labels=labeld,colors=colors,autopct="%0.00f%%",shadow=True,explode=explode)
```

```python
#sales and Data visualization
import pandas as pd

sdata=pd.read_csv("salesdata.csv")
print(sdata)
```

```python
sdata.info
totalsales_data=sdata.groupby('Month')['Total Units']
for x in totalsales_data:
    print(x)

total=totalsales_data=sdata.groupby('Month')['Total Units'].sum()
print(total)
```

```python
#plotting bar graph

total.plot(kind="bar",color='blue')
```

```python
sales_trend=sdata.groupby('Total Units')['Total Profit'].sum().reset_index()
print(sales_trend)
plt.plot(sales_trend['Total Units'],sales_trend['Total Profit'])
```
