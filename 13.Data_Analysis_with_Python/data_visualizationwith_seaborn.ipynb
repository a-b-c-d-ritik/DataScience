```python
import seaborn as sb;
```

```python
#basic plotting with seaborn
tipsd=sb.load_dataset('tips')#no need of .csv extension just name is enough
```

```python
#scatterplot
sb.scatterplot(x='total_bill',y='tip',data=tipsd)
```

```python
#lineplot

sb.lineplot(x='total_bill',y='tip',data=tipsd)
```

```python
#categorical 
#fields=categorical variable

#bar plot
sb.barplot(y='total_bill',x='day',data=tipsd)
```

```python
sb.barplot(x='total_bill',y='tip',data=tipsd)
#title dalne k liye matplotlib use kre
```

```python
#box plot

sb.boxenplot(x='total_bill',y='tip',data=tipsd)
```

```python
sb.boxplot(y='total_bill',x='day',data=tipsd)
```

```python
#voilon  plot

sb.violinplot(y='total_bill',x='day',data=tipsd)
```

```python
#histograms
sb.histplot(tipsd['day'],bins=10,kde=True)#kde=kernel density estimator
```

```python
#KDE plot
sb.kdeplot(tipsd['total_bill'],shade=True)#shade is same as fill
```

```python
#pairplot

sb.pairplot(tipsd)
```

```python
#HEatmap

correlation=tipsd[['total_bill','day','size']].corr()
#sb.heatmap(correlation,annot=True,cmap='coolwarm')
print(correlation)
```
