```python
import pandas as pd;
```

```python
print(pd)
df=pd.read_csv('dataset.csv')
#fetch first 5 rows
print(df.head(5))
print(df.describe())
```

```python
print(df.dtypes)
print(df.isnull())
print(df.isnull().sum())
print(df.isnull().sum(axis=1))
```

```python
df_filed=df.fillna(0)
```

```python
#print(df.isnull())
df['fillNA']=df['mileage'].fillna(df['mileage'].mean())
print(df.isnull())
```

```python
#rename coloumn
df=df.rename(columns={'cylinders':'cars cylinder'})
print(df.isna())
```

```python
#chnaging datatype
df['new col']=df['cars cylinder'].fillna(0).astype(int)
#df.isna()
df.head(5)
```

```python
#applying fn's
df['mileage']=df['mileage'].apply(lambda x:x*2)
df.head(2)
```

```python
#group
groupped_mean=df.groupby('mileage')['new col'].mean()
print(groupped_mean)
```

```python
group_sum=df.groupby(['new col','mileage'])['mileage'].sum()
print(group_sum)
```

```python
#aggregate multiple functions
group_agg=df.groupby(['new col','mileage'])['mileage'].agg(['mean','sum','count'])
print(group_agg)
```
