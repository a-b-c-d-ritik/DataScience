```python
from sklearn.datasets import make_classification
```

```python
new_data=make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12,n_redundant=0)
```

```python
print(new_data)
#x,y=new_data
x,y=make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12,n_redundant=0)
```

```python
#print(x,"\n",y)
#bekar h aise independent or dependent m split nhi hoga
```

```python
import pandas as pd
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df
```

```python
final_df['target'].value_counts()
# x!=y so imbalanced dataset
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```
